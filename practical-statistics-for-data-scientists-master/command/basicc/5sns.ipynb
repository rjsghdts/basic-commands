```python
import seaborn as sns
df = sns.load_dataset('tips')
df
```

## Distribution Plot
분포 시각화

```python
# Hist Plot
# 변수에 대한 히스토그램 표시, 하나 혹은 두개의 변수 분포를 나타내는 시각화 도구
sns.histplot(x=df['total_bill'])
sns.displot(x=df['total_bill'],y=df['tip'],kind='hist')
```

```python
# dist plot
# rug, kde표시 기능있음  hist보다 자주 사용함
iris=sns.load_dataset('iris'); x=iris.petal_length.values
sns.distplot(x,kde=True,rug=True)
```

```python
# KDE Plot 커널밀도추정과 관련
# 하나 혹은 두 개의 변수에 대한 분포, 밀도 추정치의 시각화, 연속된 곡선의 그래프
sns.kdeplot(x=df['total_bill'])
# 겹쳐서 나오네 ㅅㅂ
sns.kdeplot(x=df['total_bill'],y=df['tip'])
```

```python
# ECDF Plot
# 누적 분포를 시각화, 실제 관측치의 비율을 시각화, 단일 변수만 가능한듯
sns.ecdfplot(x=df['total_bill'])
```

```python
# Rug Plot
# x,y 축을 따라 눈금을 그려 주변 분포도 표시, 개발 관측치에 대한 위치를 보여 다른 그림들을 보완
sns.kdeplot(x=df['total_bill'])
sns.rugplot(x=df['total_bill']) # 주황선이 rugplot 보조 그래프 정도로 쓰임
```

## Categorical Plot
범주형 시각화 - 일변량 | 이변량 분석

```python
# Bar Plot
# 이변량 분석을 위한 plot. x축은 범주형 변수, y축은 연속형 변수
sns.barplot(x=df['sex'],y=df['tip'])
```

```python
# Count Plot
# 범주형 변수의 발생 횟수를 샘. 일변량 분석
sns.countplot(x=df['sex'])
```

```python
# Box Plot
# max,min,mean,first quartile, third quartile 관찰 용이, 단일 연속형 변수 수치 표시, 연속형 변수 기반 범주형 변수분석
sns.boxplot(x=df['total_bill'])
sns.boxplot(x=df['smoker'],y=df['total_bill'])
```

```python
# Violin Plot
# 상자그림과 유사 분포에 대한 보충정보 제공
sns.violinplot(x=df['smoker'],y=df['total_bill'])
```

```python
# Strip Plot
# 연속형과 범주형의 사이, 산점도로 표시
sns.stripplot(x=df['smoker'],y=df['total_bill'])
```

```python
# Swarm Plot
# strip과 violin의 조합
sns.swarmplot(x=df['smoker'],y=df['total_bill'])
```

## Matrix Plot
2차원 행렬 데이터 사용

```python
# Heat Map
# 데이터 간의 수치에 따라 색을 입혀 직관적인 시각화
df.corr()
sns.heatmap(df.corr(),annot=True,cmap='viridis')
```

```python
# Cluster Map
# 몇몇 특징들을 그룹화
sns.clustermap(df.corr(),annot=True,cmap='viridis')
```

## Multi plot Grid
시각화 제어력 향상 및 다양한 그래프 표시

```python
# Facet Grid
# 조건에 따라 그래프 각각 확인시 사용
sns.FacetGrid(df, col='smoker', row='sex').map(sns.distplot,'total_bill')
# 흡연 여부, 성별에 따라 4개의 그래프 형성됨
```

```python
# Joint Plot
# 두 변수에 대한 displot의 조합, 두 변수에 분포에 대한 분석
sns.jointplot(x=df['total_bill'],y=df['tip'],kind='scatter')
sns.jointplot(x=df['total_bill'],y=df['tip'],kind='hex')
```

## Regression Plot
회귀 결과 표현

```python
# Reg Plot
# 회귀 결과 그래프
sns.regplot(x='tip',y='total_bill',data=df)
```

```python
# LM Plot
# regplot 과 facegrid 결합, hue에 들어간 컬럼의 값을 구분하여 따로따로 모델링하여 결과 표현
sns.lmplot(x='tip',y='total_bill',data=df,hue='sex')
```
