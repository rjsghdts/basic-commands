### 시각화 패키지 Matplotlib | 서브 패키지 pyplot

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

### 기본설정

```python
# 폰트 설정
font1 = {'size':24, 'color':'blue'}
# 기본 설정
plt.title("Plot Name")
plt.plot([10,20,30,40],[1,4,9,16]) 
plt.xlabel("x label name", fontdict=font1)
plt.ylabel("y label name")
plt.show()
```

```python
# 스타일 지정
# 색, 마커, 선종류 순으로 지정. 생략시 디폴트 값 적용
plt.plot([10,20,30,40],[1,4,9,16],c='b',lw=5,ls='--',marker='o',ms=15,mec='g',mew=5,mfc='r')
# c 선 색, lw 선 굵기, ls 선 스타일, marker 마커종류, ms 마커크기, mec 마커 선 색
# mew 마커 선 굵기, mfc 마커 내부 색
plt.title("ex")
plt.show()
```

```python
# 축 범위 지정
plt.title("x,y range set")
plt.plot([10,20,30,40],[1,4,9,16],c='b',marker='.',ms=10)
plt.xlim(0,50)
plt.ylim(-10,30)
plt.show()
```

```python
# 틱 설정
import numpy as np
import matplotlib.pyplot as plt
X=np.linspace(-np.pi, np.pi, 256)
C=np.cos(X)
plt.title('x,y tick label setting')
plt.plot(X,C)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.yticks([-1,0,1],["Low","Zero","High"])
#그리드 설정
plt.grid(True)
plt.show()
```

```python
# 여러개의 선을 그리기
t=np.arange(0.,5.,0.2)
plt.title('many lines')
plt.plot(t,t,'r--',t,0.5*t**2,'bs:',t,0.2*t**3,'g^-')
plt.show()
```

```python
# 겹쳐 그리기, 범례 설정, 라벨타이틀
plt.title('many plot in one fig')
plt.plot([1,4,9,16],'b--o',label='blue line')
plt.plot([9,16,4,1],'r--o',label='red line')
# 라벨타이틀 설정
plt.xlabel('time')
plt.ylabel('value')
# 범례 위치설정
plt.legend(loc=2)
plt.show()
```

```python
# 구조 - Fig
# Fig(그림),Axes(플롯),Axis(축) 으로 구성
# figure 명령어는 여러개의 윈도우를 동시에 띄워야 할 때 사용
np.random.seed(0)
f1=plt.figure(figsize=(10,2))
plt.title('figure size: (10,2)')
plt.plot(np.random.randn(100))
plt.show()
```

```python
# 구조 - Axe, subplot
# 하나의 f안에 여러개의 플롯 배열시 필요
# fig를 행렬, axe를 행렬의 원소라고 생각. 위 아래 2개의 플롯이 있는 경우 2행 1열
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)
y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)

ax1=plt.subplot(2,1,1)
plt.plot(x1,y1,'yo--')
ax2=plt.subplot(2,1,2)
plt.plot(x2,y2,'ro--')
plt.tight_layout()
plt.show()
```

```python
# 복수의 axe객체를 동시에 생성도 가능
fig, axes = plt.subplots(2, 2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
# 하나의 Axe(요소)는 두 개 이상의 Axis(축)을 가진다
fig, ax0 = plt.subplots()
ax1=ax0.twinx()
ax0.set_title('2개의 y축을 한 fig에서 사용')
ax0.plot([10,5,2,9,7],'r-',label='y0')
ax0.set_ylabel('y0')
ax0.grid(False)
ax1.plot([100,200,220,180,120],'g:',label='y1')
ax1.grid(False)
ax0.set_xlabel('공유 x 축')
plt.show()
```

### 여러가지 플롯

```python
# bar chart - bar, barh(가로방향)
y=[2,3,1]
x=np.arange(len(y))
xlabel=['a','b','c']
plt.title('Bar chart')
plt.bar(x,y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel('abc')
plt.ylabel('frequency')
plt.show()

# error bar 추가 - xerr/yerr
np.random.seed(0)
people=['a','b','c','d']
y_pos=np.arange(len(people))
performance=3+10*np.random.rand(len(people))
error= np.random.rand(len(people))

plt.title('Barh Chart')
plt.barh(y_pos,performance,xerr=error,alpha=0.4) # alpha는 투명도 나타냄 0~1 값
plt.yticks(y_pos,people)
plt.xlabel('x label')
plt.show()
```

```python
# 스템 플롯 - 폭이 없는 바 차트 - 이산 확률 함수, 자기상관관계를 묘사할 때 사용
x=np.linspace(0.1,2*np.pi,10)
plt.title('Stem Plot')
plt.stem(x,np.cos(x),'-.')
plt.show()
```

```python
# 파이 차트 - 카테고리 별 값의 상대적인 비교 (솔직히 잘 안씀)
plt.axis('equal')
sizes=[15,30,45,10]
labels=['frog','pig','dog','tree']
colors=['yellowgreen','gold','lightskyblue','lightcoral']
explode=(0,0.1,0,0)
plt.title('Pie Chart')
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.show()
```

```python
# 히스토그램 - hist명령으로 bins(구간)인수를 받음
np.random.seed(0)
x=np.random.randn(1000)
plt.title('Histogram')
arrays,bins,patches=plt.hist(x,bins=10)
plt.show()
```

```python
# 스캐터 플롯(산점도) 2차원 데이터의 상관관계 알아보는 방법
np.random.seed(0)
X=np.random.normal(0,1,100)
Y=np.random.normal(0,1,100)
plt.title('Scatter Plot')
a=plt.scatter(X,Y)
plt.show()

# 3차원 이상일 때 색 혹은 점의 크기로 비교가능 - bubble chart
N=30
np.random.seed(9)
x=np.random.rand(N)
y1=np.random.rand(N)
y2=np.random.rand(N)
y3=np.pi*(15*np.random.rand(N))**2
plt.title("Bubble Chart")
plt.scatter(x,y1,c=y2,s=y3) # c - 색, s - 크기
plt.show()
```

```python
# Imshow(행렬 형태의 2차원 데이터 크기를 색으로 표시)
from sklearn.datasets import load_digits
digits=load_digits()
X=digits.images[0]
X
plt.title('mnist digits; 0')
plt.imshow(X,interpolation='nearest',cmap=plt.cm.bone_r) 
# interpolation - 표현 형식인듯? | cmap=plt.cm.~ 데이터를 색으로 바꿈
plt.xticks([])
plt.yticks([])
plt.grid(False)
plt.subplots_adjust(left=0.35,right=0.65,bottom=0.35,top=0.65)
plt.show()

methods =[None,'none','nearest','bilinear','bicubic','spline16','spline36',
        'hanning','hamming','hermite','kaiser','quadric','catrom','gaussian',
        'bessel','mitchell','sinc','lanczos']
fig, axes = plt.subplots(3,6,figsize=(12,6),subplot_kw={'xticks':[],'yticks':[]}) # subplot_kw 는 잘 모르겠음
for ax, interp_method in zip(axes.flat,methods): # axes.flat 에서 flat을 꼭 써야되는건가? 역할이 뭐지
    ax.imshow(X,cmap=plt.cm.bone_r,interpolation=interp_method)
    ax.set_title(interp_method)
plt.show()
```

```python
# 컨투어 플롯 - 입력 변수가 두 개이고 출력 변수가 하나인 경우(3차원자료)를 시각화함(등고선 형식)
# contour - 등고선
# contourf - 등고선+색 사용, 
# 입력 변수는 그대로 사용불가 하고 meshgrid 명령으로 그리드 포인트 행렬 만들어야 함
# contour(x,y,z) z에 들어있는 x,y좌표 지정
# contour(~,levels) 열거된 구문에서 표시할 등고선 지정
# contour(~,LineSpec) 등고선 스타일과 색 지정
# [M,c]=contour() 등고선행렬M과 contour객체 c반환. c.~로 사용가능해짐
def f(x,y):
    return(1-x/2+x**5+y**3)*np.exp(-x**2 -y**2)
n=256
x=np.linspace(-3,3,n)
y=np.linspace(-3,3,n)
XX,YY=np.meshgrid(x,y)
ZZ=f(XX,YY)
plt.title("Contour plots")
plt.contourf(XX,YY,ZZ,alpha=.75,cmap='jet')
plt.contour(XX,YY,ZZ,colors='black')
plt.show()
```

```python
# 3D 서피스 플롯
# plot_wireframe,plot_surface 사용
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
XX,YY=np.meshgrid(X,Y)
RR=np.sqrt(XX**2+YY**2)
ZZ=np.sin(RR)
fig=plt.figure()
ax=Axes3D(fig)
ax.set_title('3D Surface Plot')
ax.plot_surface(XX,YY,ZZ,rstride=1,cstride=1,cmap='hot')
plt.show()
```

### triangular grid 사용법 | 여기는 ㄹㅇ 모르겠다 ㅋㅋ

```python
# 패키지
import matplotlib.tri as mtri
# 클래스
# Triangulation -  삼각 그리드 형성 | TriRefiner/UniformTriRefiner - 삼각그리드 세분화
# triplot,tricontour,tricontourf,tripcolor - 삼각 그리드 플롯
# TriInterpolator, LinearTriInterpolator, CubicTriInterpolator - 삼각 그리드 보간
```

```python
# 삼각 그리드 생성 - x,y,triangles 세개의 인수를 받음 triangles가 주어지지 않으면 자동 생성
x=np.array([0,1,2])
y=np.array([0,np.sqrt(3),0])
triangles=[[0,1,2]]
triang=mtri.Triangulation(x,y,triangles)
plt.title('삼각그리드')
plt.triplot(triang,'ko-')
plt.xlim(-0.1,2.1)
plt.ylim(-0.1,1.8)
plt.show()

x = np.asarray([0, 1, 2, 3, 4, 2])
y = np.asarray([0, np.sqrt(3), 0, np.sqrt(3), 0, 2*np.sqrt(3)])
triangles = [[0, 1, 2], [2, 3, 4], [1, 2, 3], [1, 3, 5]]
triang = mtri.Triangulation(x, y, triangles)
plt.title("여러개의 삼각형 그리기")
plt.triplot(triang, 'ko-')
plt.xlim(-0.1, 4.1)
plt.ylim(-0.1, 3.7)
plt.show()
```

```python
# 그리드 세분화
refiner = mtri.UniformTriRefiner(triang)
triang2 = refiner.refine_triangulation(subdiv=2)
plt.title("그리드 세분화")
plt.triplot(triang2, 'ko-')
plt.xlim(-0.1, 4.1)
plt.ylim(-0.1, 3.7)
plt.show()
```

```python
# 그리드 플롯
triang5 = refiner.refine_triangulation(subdiv=5)
z5 = np.cos(1.5*triang5.x)*np.cos(1.5*triang5.y)
plt.title("삼각 그리드에서의 3차원 데이터 시각화")
plt.tricontourf(triang5, z5, cmap="gray")
plt.show()

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

triang3 = refiner.refine_triangulation(subdiv=3)
z3 = np.cos(1.5 * triang3.x) * np.cos(1.5 * triang3.y)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_title("삼각 그리드에서의 3D Surface Plot")
ax.plot_trisurf(triang3.x, triang3.y, z3, cmap=cm.jet, linewidth=0.2)
ax.tricontourf(triang3, z3, zdir='z', offset=-1.2, cmap=cm.coolwarm)
ax.set_zlim(-1, 1)
ax.view_init(40, -40)
plt.show()
```
